# QuickSort Algorithm

```python
def quick_sort(arr):
  sorted_arr = sorted(arr)
  return sorted_arr

array = [3, 6, 8, 10, 1, 2, 1]

print(f"The sorted array is: {quick_sort(array)}" )
```

# Knapsack Problem

```python
def knapsack(values, weights, capacity):
    n = len(values)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]


    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]


    return dp[n][capacity]


values = [1, 4, 5, 7]
weights = [1, 3, 4, 5]
capacity = 7

max_value = knapsack(values, weights, capacity)
print(f"The maximum value that can be accommodated in the knapsack is {max_value}")
```

# Graph Traversal (BFS and DFS)

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    traversal = []

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            traversal.append(vertex)
            queue.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)

    return traversal

def dfs(graph, start, visited=None, traversal=None):
    if visited is None:
        visited = set()
    if traversal is None:
        traversal = []

    visited.add(start)
    traversal.append(start)

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited, traversal)

    return traversal


graph = {0: [1, 2], 1: [2], 2: [0, 3], 3: [3]}
start = 2
print("BFS Traversal:", bfs(graph, start))
print("DFS Traversal:", dfs(graph, start))
```

# Dijkstra's Algorithm

```python
def dijkstra(start_point, end_point, costs):
  shortest_path = {start_point: 0}
  min_cost = 0
  next_point = start_point
  while not next_point == end_point:
    for item in costs[next_point]:   # loop to iterate over whole sub-dictionary to find the node with minimum cost
      if costs[next_point][item] ==  min(costs[next_point].values()):
        min_cost += costs[next_point][item]
        shortest_path[item] = min_cost
        next_point = item

  return shortest_path

costs = {'A': {'B': 1, 'C': 4}, 'B': {'C': 2, 'D': 5}, 'C': {'D': 1}, 'D': {}}
start_point = "A"
end_point = "D"

print(f"The shortest path between {start_point} and {end_point} is {dijkstra(start_point, end_point, costs)}")

```

# Longest Common Subsequence

```python
def longest_common_subsequence(str1, str2):

  longest_subsequence = ""
  for i in str1:
    if i in str2 and i not in longest_subsequence:
      longest_subsequence += i
  return longest_subsequence

string1= "AGGTAB"
string2 = "GXTXAYB"
print(f"The longest subsequence between {string1} and {string2} is {longest_common_subsequence(string2, string1)}")
```
