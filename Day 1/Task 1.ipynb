# Area of a Circle

```python
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * (radius **2)
print(f"The area of the circle is: {area}")
```

# Even or Odd

```python
number = int(input("Enter a number:"))
if number % 2:
  print(f"{number} is odd")
else:
  print(f"{number} is even")
```

# Simple Calculator

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /)")
if operation == "+":
  print(f"The result is: {num1 + num2}")
elif operation == "-":
  print(f"The result is: {num1 - num2}")
if operation == "*":
  print(f"The result is: {num1 * num2}")
else:
  if num2 != 0:
    print(f"The result is: {num1 / num2}")
  else:
    print("Can't divide by zero.")


```

# Find the Largest Number

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))
if num1 > num2 and num1 > num3:
  print(f"The largest number is: {num1}")
elif num2 > num1 and num2 > num3:
  print(f"The largest number is: {num2}")
else:
  print(f"The largest number is: {num3}")
```

# Count Vowels

```python
word = input("Enter a string: ")
vowel_count = 0
vowels = ['a', 'e', 'i', 'o', 'u']
for letter in word.lower():
  if letter in vowels:
    vowel_count += 1
print(f"The number of vowels is: {vowel_count}")
```

# Fibonacci Sequence

```python
previous_no = 0
current_no = 1
total_count = int(input("How many terms? "))
n_count = 0
print("Fibonacci sequence:")
while n_count < total_count:
  print(previous_no, end=" ")
  next_number = previous_no + current_no
  previous_no = current_no
  current_no = next_number
  n_count += 1
```

# Check for Palindrome

```python
def is_palindrome(word):
  if word[::-1] == word:
    print(f"{word} is a palindrome")
  else:
    print(f"{word} is not a palindrome")
is_palindrome("radar")
is_palindrome("vowel")
```
