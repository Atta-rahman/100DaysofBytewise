# Palindrome Checker

```python
def is_palindrome(word):
  if word[::-1] == word:
    print(f"{word} is a palindrome")
  else:
    print(f"{word} is not a palindrome")
is_palindrome("racecar")
is_palindrome("python")
```

# Nth Fibonacci Number

```python
def nth_fibonacci(position):
  if position == 0:
    return 0
  elif position == 1:
    return 1
  elif position == 2:
    return 1
  else:
    return nth_fibonacci(position - 1) + nth_fibonacci(position - 2)   #recursively call the function until 0 or 1 is obtained
position = int(input("Enter the position of the number:"))
nth_number = nth_fibonacci(10)
print(f"The {position}th fibonacci number is {nth_number}")
```

# Prime Number Checker

```python
def is_prime(num):
  prime = True
  for i in range(2, num):
    if num % i == 0:
      prime = False
      break
  if  prime:
    print(f"{num} is a prime number")
  else:
    print(f"{num} is not a prime number")

is_prime(59)
is_prime(96)
```

# Guess the Number Game

```python
import random
original_num = random.randint(1, 100)
while True:
  guessed_num = int(input("Enter your guess: "))
  if guessed_num > original_num:
    print("Too High!")
  elif guessed_num < original_num:
    print("Too Low!")
  else:
    print("Correct")
    break

```

# Palindrome Sentences

program to check if a given sentence is a palindrome (reads the same forwards and backwards, ignoring spaces and punctuation)

```python
import string
def is_palindrome(sentence):
  cleaned_string = ''.join( i for i in sentence.lower() if i not in string.punctuation + ' ')
  if cleaned_string[::-1] == cleaned_string:
    print(f"{sentence} is a palindrome" )
  else:
    print(f"{sentence} is not a palindrome" )

is_palindrome("A man, a plan, a canal: Panama")
is_palindrome("Hello, world!")
```

# Anagram Checker

Check if two strings are anagrams(contain the same characters in a different order).


```python
def are_anagrams(str1, str2):
  if sorted(str1.lower()) == sorted(str2.lower()):
    print(f"{str1} and {str2} are anagrams")
  else:
    print(f"{str1} and {str2} are not anagrams")

are_anagrams( "listen", "silent")
are_anagrams("python", "java")
```

# Reverse Words in a Sentence

```python
def reverse_sentence(sentence):
  reversed_sentence = ' '.join(sentence.split()[::-1])
  print(reversed_sentence)

reverse_sentence("The quick brown fox jumps over the lazy dog.")
```

# Temperature Converter

Given the value of temperature in one scale, gives its value in other scales.

```python
def temperature_converter(temperature, scale):
  if scale.lower() == "celsius":
    temperature_fahrenheit = round(1.8 * temperature + 32, 2)
    temperature_kelvin = temperature + 273.15
    print(f"{temperature} degrees Celsius is equal to {temperature_fahrenheit} degrees Fahrenheit and {temperature_kelvin} degrees Kelvin.")

  elif scale.lower() == "fahrenheit":
    temperature_celsius = round((temperature - 32) / 1.8, 2)
    temperature_kelvin = temperature_celsius + 273.15
    print(f"{temperature} degrees Fahrenheit is equal to {temperature_celsius} degrees Celsius and {temperature_kelvin} degrees Kelvin.")

  else:
    temperature_celsius = round(temperature - 273.15, 2)
    temperature_fahrenheit = round(1.8 * temperature_celsius + 32, 2)
    print(f"{temperature} degrees Kelvin is equal to {temperature_celsius} degrees Celsius and {temperature_fahrenheit} degrees Fahrenheit.")

temperature_converter(20, "Celsius")
temperature_converter(98.6, "Fahrenheit")
temperature_converter(300, "Kelvin")
```
